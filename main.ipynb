## Actividad uno

```python
import pandas as pd
df=pd.read_csv("ventas.csv")
df_backup=df
df_backup.head()
```

## Actividad dos

```python
# Mostrar los datos
print(f"Mostrar primeros 5: \n {df_backup.head()}")
print(f"Mostrar últimos 5: \n {df_backup.tail()}")
```

```python
# Ver la información detallada del DF
print(df_backup.isnull().sum())
print(df_backup.info())
```

## Actividad tres

```python
# Modificar los datos nulos utilizando el metodo fillna

df_backup["Precio_Unitario"]=df_backup["Precio_Unitario"].fillna(0)
df_backup["Total_Venta"]=df_backup["Total_Venta"].fillna(0)

df_backup["Producto"]=df_backup["Producto"].fillna(0)
df_backup["Producto"]=df_backup["Producto"].replace(0.0, "Sin_Producto") # se modificó los datos de productos que estaban en 0 a "Sin_Producto"

df_backup["Metodo_Pago"]=df_backup["Metodo_Pago"].fillna(0)
df_backup["Metodo_Pago"]=df_backup["Metodo_Pago"].replace(0.0, "Sin_Metodo") # se hizo el mismo tratamiento que para la columna de Producto

df_backup["Region"]=df_backup["Region"].fillna(0)
df_backup["Region"]=df_backup["Region"].replace(0.0, "Sin_Region") # mismo tratamiento
```

```python
# Capitalización de los campos de texto
df_backup["Nombre_Cliente"]=df_backup["Nombre_Cliente"].str.upper().str.title().str.strip()
df_backup["Apellido_Cliente"]=df_backup["Apellido_Cliente"].str.upper().str.title().str.strip()
```

```python
# Se realizó la transformación de dato a datetime
fechas=df_backup["Fecha"]
fecha_uno=df_backup["Fecha"]=pd.to_datetime(fechas, format="%Y-%m-%d", errors="coerce")
```

## Actividad cuatro


```python
df_backup.describe()
```

```python
#Agrupamos dos columnas (Region y metodo de pago) y mostramos el total de ventas que hubo en esos lugares
df_backup.groupby(["Region","Metodo_Pago"])["Total_Venta"].sum()
```

# Gráfico con Matplotlib
Se realizó la importación de la librería de Matplotlib para poder graficar

```python
import matplotlib.pyplot as plt

metodos=df_backup.groupby("Region")["Total_Venta"].sum()

plt.figure(figsize=(20,5))
plt.title("Total de Ventas por Region")

plt.plot(metodos.index, metodos.values)
plt.xlabel("Region")
plt.ylabel("Cantidad de Ventas")

plt.grid()
plt.show()
```

## Exportación de archivo a SQLite
Se exportó un nuevo archivo utilizando la librería SQLite para usarlo como una base de datos

```python
import sqlite3
conn=sqlite3.connect("ventas.db")
df_backup.to_sql("ventas", conn, if_exists="replace", index=False)
pd.read_sql("SELECT * FROM Ventas", conn)

conn.close()
```
